## Question 1 Population database.

```python
import sqlite3
# Setting the minimum and maximum as well as the different options for my display menu.
MIN_CHOICE = 1
MAX_CHOICE = 8
CITIES_ASCENDING = 1
CITIES_DESCENDING = 2
CITIES_LIST = 3
CITIES_TOTAL = 4
CITIES_AVERAGE = 5
CITIES_HIGHEST = 6
CITIES_LOWEST = 7
EXIT = 8
# The start of my program.
def main():
    # Sets choice to 0
    choice = 0
    # While the users choice doesn't equal EXIT it will run both display_menu() and choice = get_menu_choice()
    while choice != EXIT:
        display_menu()
        choice = get_menu_choice()
        # if elif to preform whichever the user chooses.
        if choice == CITIES_ASCENDING:
            cities_ascending()
        elif choice == CITIES_DESCENDING:
            cities_descending()
        elif choice == CITIES_LIST:
            cities_list()
        elif choice == CITIES_TOTAL:
            cities_total()
        elif choice == CITIES_AVERAGE:
            cities_average()
        elif choice == CITIES_HIGHEST:
            cities_highest()
        elif choice == CITIES_LOWEST:
            cities_lowest()
# assigning what display_menu() will do which is a bunch of print functions.            
def display_menu():
    print('\n----- Population Database Menu -----')
    print('1. List cities population ascending')
    print('2. List cities population descending')
    print('3. List cities')
    print('4. List total population of all cities')
    print('5. List average population of cities')
    print('6. List the highest population city')
    print('7. List the lowest population city')
    print('8. Exit the program')
# Assigns the user input and a validation that if an incorrect input will ask them to input again.    
def get_menu_choice():
    choice = int(input('Enter your choice: '))
    # validation for min and max input.
    while choice < MIN_CHOICE or choice > MAX_CHOICE:
        print(f'Valid choices are {MIN_CHOICE} through {MAX_CHOICE}.')
        choice = int(input('Enter your choice: '))
        
    return choice
# 
def cities_ascending():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT * FROM CityName ORDER BY Population')
    ascending = cur.fetchall()
    for row in ascending:
        print(f'{row[1]:30} {row[2]:10})
    conn.close()
    
def cities_descending():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT * FROM CityName ORDER BY Population DESC')
    descending = cur.fetchall()
    for row in descending:
        print(f'{row[1]:30} {row[2]:10})
    conn.close()
    
def cities_list():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT COUNT FROM CityName')
    city_list = cur.fetchall()
    for row in city_list:
        print(f'{row[1]:30})
    conn.close()
    
def cities_total():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT SUM(Population) FROM CityName')
    total = cur.fetchall()
    for row in total:
        print(f'{row[2]:10})
    conn.close()
    
def cities_average():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT AVG(Population) FROM CityName')
    average = cur.fetchall()
    for row in average:
        print(f'{row[2]:10})
    conn.close()
    
def cities_highest():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT MAX(Population) FROM CityName')
    highest = cur.fetchall()
    for row in highest:
        print(f'{row[1]:30} {row[2]:10})
    conn.close()
    
def cities_lowest():
    conn = sqlite3.connect('cities.db')
    cur = conn.cursor()
    cur.execute('SELECT MIN(Population) FROM CityName')
    lowest = cur.fetchall()
    for row in lowest:
        print(f'{row[1]:30} {row[2]:10})
    conn.close()

main()
    
    
```
